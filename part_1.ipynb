````python
# return_odd_or_even_phrase takes in an int and returns a phrase indicating if the number is odd or even. It uses the modulo and equality operators to assess if the number is odd or even
def return_odd_or_even_phrase(num: int) -> str:
    return f"this number is {'even' if num % 2 == 0 else 'odd'}"
````

ii) Using a different statement, identify if the date variable you defined in task 1 occurs either before 1st
January 2000, between 1st January 2000 and 31st December 2009, or after 31st December 2009.

````python
# I choose to include type hints. How they affect readability is arguable, but they certainly fit the  2nd zen of Python - "Explicit is better than implicit."
# For the purpose of this exercise, a single date input would have been best but I lean toward re-usability, so I have allowed for date range args with the default set to the brief's date settings.
def is_date_before_during_or_after_range(date_to_test: date, start_of_range: date = date(2000, 1, 1), end_of_range: date = date(2009, 12, 31)) -> str:
    if start_of_range > end_of_range: # Arguably the end date should be not be allowed to be the same as the start date
        raise ValueError("start_of_range must be before or equal to end_of_range")
    if date_to_test < start_of_range:
        return f"{date_to_test} is before the specified range, {start_of_range} to {end_of_range}."
    if date_to_test > end_of_range:
        return f"{date_to_test} is after the specified range, {start_of_range} to {end_of_range}."
    else:
        return f"{date_to_test} is during the specified range, {start_of_range} to {end_of_range}."
# If taken to the letter, the brief does not account for dates that are the same as the start or end of the date range (before, after or between) but I have chosen to consider the between as inclusive rather than handle these separately as this suits real-world usage.
````

````python
print(is_date_before_during_or_after_range(release_date)) # Output: 2009-05-27 is during the specified range, 2000-01-01 to 2009-12-31.
assert is_date_before_during_or_after_range(release_date) == "2009-05-27 is during the specified range, 2000-01-01 to 2009-12-31." # Note that assert does not require the optional error statement. The error that would be thrown if the condition is ot true can be checked by deliberately making the condition false.
````

The third part of this section, iii, does not really make sense unless the values are numbers, but it does make it more interesting with different types.

Multiplying a list by 2, repeats the existing items inside the list.

Technically, I probably should have left the strings handled the same way, but chose to add a space between the doubled strings. It formats nicely though does not really meet the requirements. My justification for this was that it demonstrates an understanding of what would happen AND its limitations, i.e. poor formatting, and so, neither would be quite right but this way achieves more in the wider brief of demonstrating understanding and context.

Date was interesting. You can't really double a date unless there is a point at which 'date' begins. I decided to use 0AD or 0BC but date(0, 0, 0) is not possible. I imagine I could have found a workaround but I decided to go with a more secular and no doubt equally controversial 6th July 1189 as the date starting point as this is the date which English law considers to be the limit of legal memory, 'time immemorial'.

Runtime, 1h 56m, was also an interesting case. It would be simple enough to deconstruct the string, calculate the number of minutes, double it and then reconstruct an hour and minutes string, but this seemed too obtuse, so I simply treated it as the string that it is.

NOTE: It would probably be better to do this with a simple numerical dict, but since there is no functional purpose to this doubling, this seems fine. Booleans are noticeably unhandled since there was not any in dict.

````python
doubled_values = []
for value in film_1.values(): # Loop over the values in the dict
    if isinstance(value, (int, float, list)): # if value is any of these types
        doubled_values.append(value * 2)
    elif isinstance(value, str):
        doubled_values.append(f"{value} {value}")
    elif isinstance(value, date):
        year_one = date(1189, 7, 6)
        delta = (value - year_one).days
        doubled_date = year_one + timedelta(days=delta * 2)
        doubled_values.append(doubled_date)

assert doubled_values[0] == 'Looking for Eric Looking for Eric', f"Unexpected string value: {doubled_values[0]}"
assert doubled_values[2][0] == 'Steve Evets', f"Unexpected list item string value: {doubled_values[2][0]}"
assert doubled_values[4] == 14.2, f"Unexpected float value: {doubled_values[4]}"
assert doubled_values[5] == 4018, f"Unexpected int value: {doubled_values[5]}"
assert doubled_values[6] == date(2829, 3, 30), f"Unexpected date value: {doubled_values[6]}"
print(f"doubled_values: {doubled_values}") # Output: doubled_values: ['Looking for Eric Looking for Eric', 'Ken Loach Ken Loach', ['Steve Evets', 'Éric Cantona', 'Stephanie Bishop', 'Steve Evets', 'Éric Cantona', 'Stephanie Bishop'], '1h 56m 1h 56m', 14.2, 4018, datetime.date(4017, 10, 2), datetime.date(2829, 3, 30), 6, 12]
````

````python
import numpy as np
````

````python
numbers = [1, 3, 7, 13, 21] # A list of numbers
numbers_array = np.array(numbers) # Turn the list into a NumPy array which stores the numbers efficiently in memory, forces all items to be the same type and allows fast mathematical operations on the whole array.
````

````python
np_mean = np.mean(numbers_array)
np_median = np.median(numbers_array)
np_population_standard_deviation = np.std(numbers_array)
np_sample_standard_deviation = np.std(numbers_array, ddof=1)

print(f"np_mean: {np_mean}")
print(f"np_median: {np_median}")
print(f"np_population_standard_deviation: {np_population_standard_deviation}")
print(f"np_sample_standard_deviation: {np_sample_standard_deviation}")
````

Creating a function to do all of this makes the process easily repeatable. It is reasonable that such a function should produce the mean, median and standard deviation, but in the real world, would not produce them in two different ways and then make assertions before returning the statistics. This would be a simple function using NumPy which might have its own unit tests elsewhere.

````python
def calculate_mean_median_and_std(numbers) -> dict:
    numbers_array = np.array(numbers)
    np_mean = np.mean(numbers_array)
    np_median = np.median(numbers_array)
    np_population_standard_deviation = np.std(numbers_array)
    np_sample_standard_deviation = np.std(numbers_array, ddof=1)

    number_series = pd.Series(numbers)
    pd_mean = number_series.mean()
    pd_median = number_series.median()
    pd_population_standard_deviation = number_series.std(ddof=0)
    pd_sample_standard_deviation = number_series.std(ddof=1)

    assert np_mean == pd_mean, generate_numpy_pandas_discrepancy_error(np_mean, pd_mean, "mean")
    assert np_median == pd_median, generate_numpy_pandas_discrepancy_error(np_median, pd_median, "median")
    assert np_population_standard_deviation == pd_population_standard_deviation, generate_numpy_pandas_discrepancy_error(np_population_standard_deviation, pd_population_standard_deviation, "population standard deviation")
    assert np_sample_standard_deviation == pd_sample_standard_deviation, generate_numpy_pandas_discrepancy_error(np_sample_standard_deviation, pd_sample_standard_deviation, "sample standard deviation")

    return {
        "mean": np_mean,
        "median": np_median,
        "pop_std": np_population_standard_deviation,
        "sample_std": np_sample_standard_deviation
    }
````

### Part 1 Task 5: User-defined functions

Create your own function that takes a string and an integer as parameters. The function should shorten the
string (from the right-hand side) by the number of characters indicated by the integer, and then return the
string. Show suitable testing of your function.

````python
def shorten_string(starting_string: str, shorten_by: int) -> str:
    if not isinstance(starting_string, str):
        raise TypeError("starting_string must be an string")
    if not isinstance(shorten_by, int) or isinstance(shorten_by, bool):
        raise TypeError("shorten_by must be an integer")
    if shorten_by < 0:
        raise ValueError("shorten_by must be non-negative")
    return starting_string[:-shorten_by] if shorten_by else starting_string
````

👆🏻  The validation for the first argument is not really necessary but it does deliver a very clear error message.

The second part of the second argument validation is necessary since True and False Boolean values will be read as 1 and 0 respectively and so, return unexpected results rather than a type error unless it is the intention of the function to accept Booleans as numerical values.

Rejecting negative shorten_by numbers is a design decision.

There is an ambiguity as to whether it is acceptable to pass in a shorten_by int that is equal to or greater than the length of starting_string. Even though "Hello" cannot be shortened by 8 characters, I decided that returning an empty string in such a scenario meets the brief better than raising an error or warning.

````python
# NOTE: unittest was imported at the top of this notebook.

# A unittest test class must inherit from unittest.TestCase.
# unittest will automatically discover and run any method in that class that starts with the word "test".
class TestShortenString(unittest.TestCase):

    def test_simple_cases(self): # self is the instance of TestShortenString instantiated by unittest.
        # Below I am using a list of cases - tuples - to parameterise the tests.
        # Each tuple is: (input_string, shorten_by_value, expected_output)
        simple_cases = [
            ("hello", 0, "hello"),
            ("hello", 2, "hel"),
            ("hello", 1, "hell"),
            ("hello", 5, ""),
            ("h", 1, ""),
        ]

        # Loop through each test case using the values deconstructed from the tuples
        for starting_string, shorten_by, expected in simple_cases:
            # subTest() creates a *separate labelled test*
            # for each set of values. If one fails, the others
            # still run, and the output shows which case failed.
            with self.subTest(starting_string=starting_string, shorten_by=shorten_by):

                # assertEqual(actual, expected)
                # Passes if the two values are equal.
                self.assertEqual(
                    shorten_string(starting_string, shorten_by),
                    expected
                )

    def test_valid_edge_cases(self):
        # Edge cases are unusual but valid inputs
        valid_edge_cases = [
            ("", 0, ""), # empty string
            ("hello", 10, ""), # shorten by more than the string length
            ("1\n2\n3", 2, "1\n2"), # newline characters inside string
            ("     ", 2, "   "), # string made only of spaces
        ]

        for starting_string, shorten_by, expected in valid_edge_cases:
            with self.subTest(starting_string=starting_string, shorten_by=shorten_by):
                self.assertEqual(
                    shorten_string(starting_string, shorten_by),
                    expected
                )

    def test_type_error_cases(self):
        # These inputs should raise TypeError
        type_error_cases = [
            (123, 2), # starting_string is not a string
            (None, 1), # starting_string is None
            ("hello", "2"), # shorten_by is not an integer
            ("hello", True), # bool is a subclass of int but is rejected in my implementation
        ]

        for starting_string, shorten_by in type_error_cases:
            with self.subTest(starting_string=starting_string, shorten_by=shorten_by):
                # assertRaises(TypeError) passes if the code inside the block raises that exception.
                with self.assertRaises(TypeError):
                    shorten_string(starting_string, shorten_by)

    def test_value_error_cases(self):
        # These inputs should raise ValueError since negative shortening is not allowed
        value_error_cases = [
            ("hello", -5),
            ("hello", -1),
        ]

        for starting_string, shorten_by in value_error_cases:
            with self.subTest(starting_string=starting_string, shorten_by=shorten_by):
                with self.assertRaises(ValueError):
                    shorten_string(starting_string, shorten_by)


# unittest.main() runs the tests.
# argv=[''] prevents Jupyter from passing its own arguments to unittest.
# exit=False stops unittest from shutting down the notebook kernel.
# defaultTest='TestShortenString' tells unittest to run only this class.
unittest.main(argv=[''], exit=False, defaultTest='TestShortenString')
````

````python
def try_shorten_string(starting_string: str, shorten_by: int) -> str | None:
    try:
        if not isinstance(starting_string, str):
            raise TypeError("starting_string must be a string")
        if not isinstance(shorten_by, int) or isinstance(shorten_by, bool):
            raise TypeError("shorten_by must be an integer")
        if shorten_by < 0:
            raise ValueError("shorten_by must be non-negative")
        return starting_string[:-shorten_by] if shorten_by else starting_string
    except (TypeError, ValueError) as error:
        print(f"ERROR: {error}")
````

👆🏻  Whereas ```shorten_string``` exits with the error, ```try_shorten_string``` 'handles' any ```TypeError``` that arises.

Which is correct depends on the intention and how the return is handled. The example above breaks the function contract to return a string in the case of a ```TypeError```.

When ```shorten_string``` is passed incorrect types:
- Exits the program with an error

When ```try_shorten_string``` is passed incorrect types:
- Allows the program to continue
- ```None``` is returned in place of the 'promised' string (now amended to ```str | None```)
- The ```None``` return must be handled correctly where the function was called

Most engineers at some point create an unbreakable test - a test that cannot fail no matter what - or one that simply does not test what you are expecting such as the example below:

````python
class TestNothing(unittest.TestCase):

    def test_nothing(self):
        simple_cases = [
            ("hello", 0, "hello"),
            ("hello", 2, "hel"),
            ("hello", 1, "hell"),
            ("hello", 5, ""),
            ("h", 1, ""),
        ]

        for starting_string, shorten_by, expected in simple_cases:
            with self.subTest(starting_string=starting_string, shorten_by=shorten_by):
                result = try_shorten_string(starting_string, shorten_by)
                self.assertNotEqual(shorten_by, result)
                self.assertEqual(len("hello"), 5)

unittest.main(argv=[''], exit=False, defaultTest='TestNothing')
````

The example is nonsense, but it highlights that just because tests run and pass, it does not mean that your code is correct unless you have also tested you tests.

An example of this is deliberate test breaking, e.g.:
- run your tests - they should pass
- then 'break' something in the test, e.g. change the case ```("hello", 0, "hello"),``` to ```("hello", 0, "hell"),```
- run the tests again - the test you 'broke' should fail
- If all the tests still passed then something smells off
