## Data types and structures

```python
from typing import runtime_checkable

# Examples of strings
film_name = "Looking for Eric"
director = "Ken Loach"

print(film_name) # Output: Looking for Eric
print(director) # Output: Ken Loach
```

```python
# Examples of floats
runtime_hours = 1.933 # Equivalent to 1 hour and 56 minutes
imbd_rating = 7.1

print(runtime_hours) # Output: 1.933
print(imbd_rating) # Output: 7.1
```

```python
# Examples of ints (integers)
year = 2009
number_of_award_nominations = 6
number_of_award_wins = 3

print(year)  # Output: 2009
print(number_of_award_nominations)  # Output: 6
print(number_of_award_wins)  # Output: 3
```

```python
# Examples of dates
premiere_date = date(2009,5,18) # Date from datetime always uses the format (year, month, day) so there is no format ambiguity.
release_date = date(2009,5,27)

print(premiere_date)  # Output: 2009-05-18
print(release_date)  # Output: 2009-05-27
```

```python
# Examples of Bools (Booleans)
was_filmed_in_manchester = True
won_palme_dor_award = False

print(was_filmed_in_manchester)  # Output: True
print(won_palme_dor_award)  # Output: False
```

```python
# Examples of lists
starring_actors = ["Steve Evets", "Éric Cantona", "Stephanie Bishop"]
films_directed_by_eric_cantona = ["Liberty, Equality,Fraternity and Football", "Foot et immigration, 100 ans d'histoire commune", "Apporte-moi ton amour"] # Note that English titles use title case whilst French titles use sentence case.

print(starring_actors)  # Output: ['Steve Evets', 'Éric Cantona', 'Stephanie Bishop']
print(films_directed_by_eric_cantona)  # Output: ['Liberty, Equality,Fraternity and Football', "Foot et immigration, 100 ans d'histoire commune", 'Apporte-moi ton amour']
```
