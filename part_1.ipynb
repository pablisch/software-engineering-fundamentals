```python
# return_odd_or_even_phrase takes in an int and returns a phrase indicating if the number is odd or even. It uses the modulo and equality operators to assess if the number is odd or even
def return_odd_or_even_phrase(num: int) -> str:
    return f"this number is {"even" if num % 2 == 0 else "odd"}"
```

```python
year_phrase = return_odd_or_even_phrase(year) # year == 2009
number_of_award_nominations_phrase = return_odd_or_even_phrase(number_of_award_nominations) # number_of_award_nominations == 6
number_of_award_wins_phrase = return_odd_or_even_phrase(number_of_award_wins) # number_of_award_wins == 3

assert year_phrase == "this number is odd", f'Expected: "this number is odd" but got: "{year_phrase}"'
assert number_of_award_nominations_phrase == "this number is even", f'Expected: "this number is even" but got: "{number_of_award_nominations_phrase}"'
assert number_of_award_wins_phrase == "this number is odd", f'Expected: "this number is odd" but got: "{number_of_award_wins_phrase}"'

print(f"year: {year}\nyear_phrase: {year_phrase}\n")
print(f"number_of_award_nominations: {number_of_award_nominations}\nnumber_of_award_nominations_phrase: {number_of_award_nominations_phrase}\n")
print(f"number_of_award_wins: {number_of_award_wins}\nnumber_of_award_wins_phrase: {number_of_award_wins_phrase}\n")
```
