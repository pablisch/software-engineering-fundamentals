```python
# Accepted coin inputs and exit shortcuts
pound_coins = ["£2", "£1"]
penny_coins = ["50p", "20p", "10p", "5p", "2p", "1p"]
exit_codes = ["quit", "exit", "q"]
```

```python
# We only want to create the inventory once.
# 'try' runs the block; if it fails, Python jumps to 'except'.
try:
    # This line just checks whether the name exists.
    inventory
# NameError means the variable name hasn't been defined yet.
except NameError:
    # inventory is a dict (dictionary) mapping each item name to another dict
    # Each inner dict stores two key/value pairs: stock (int) and price (float).
    inventory = {
        "soda": {
            "stock": 10,
            "price": 1.49,
        },
        "chips": {
            "stock": 6,
            "price": 1.99,
        },
        "candy": {
            "stock": 0,
            "price": 1.19,
        }
    }
```

```python
# Returns True if the item has at least 1 in stock
# The type hints say: item is a str and the function returns a bool.
def check_stock(item: str) -> bool: 
    # inventory[item] looks up a nested dict for that item name.
    return inventory[item]["stock"] > 0
```

```python
# Ask the user what they want and report if it's in stock
def select_item() -> tuple[str, bool]:
    # input() gives a string; lower/strip normalise the user's typing.
    item = input("What would you like to buy? ").lower().strip()
    is_in_stock = False

    # Intentionally explicit branches for beginner practice
    if item == "soda":
        is_in_stock = check_stock(item)
    # 'elif' means 'else if' — it only runs if the earlier 'if' was False
    elif item == "chips":
        is_in_stock = check_stock(item)
    elif item == "candy":
        is_in_stock = check_stock(item)
    elif item in exit_codes:
        return "quit", False
    else:
        # else runs when none of the earlier conditions matched
        print(f"Sorry, we don't keep {item} in this machine.\n")

    # 'in' checks whether the item name is a key in the inventory dict
    if item in inventory:
        if is_in_stock:
            print(f"Certainly, we have {inventory[item]['stock']} of them.\n")
        else:
            print(f"Sorry, we are out of {item} today.\n")
    # Return two values as a tuple (item name, stock status)
    return item, is_in_stock
```

```python
# Take coins until the full price is paid (or user exits) then return any change due
def accept_payment(item: str, is_in_stock: bool) -> None:
    # Only ask for payment if the item is in stock
    if is_in_stock:
        coins_paid: int = 0
        amount_paid_pence: int = 0
        # Convert pounds to whole pence to avoid float issues
        price_pence = int(round(inventory[item]["price"] * 100))

        # while repeats the block until the condition becomes False
        while amount_paid_pence < price_pence:
            remaining_pence = price_pence - amount_paid_pence
            # :.2f formats the number to 2 decimal places
            remaining_display = f"£{remaining_pence/100:.2f}"
            coin = input(
                f"Please pay {remaining_display}.\n"
                f"Please enter coin {coins_paid + 1} amount (£2, £1, 50p, 20p, 10p, 5p, 2p or 1p): "
            ).strip().lower()

            # Early return exits the function immediately
            if coin in exit_codes:
                print("Payment cancelled.\n")
                return

            # 'in' checks membership in the list of accepted coins
            if coin in pound_coins:
                # Strip the £ and turn pounds into pence
                coin_value_pence = int(coin.replace("£", "")) * 100
                amount_paid_pence += coin_value_pence
                # += is an increment: add 1 to the current value
                coins_paid += 1
            elif coin in penny_coins:
                # Strip the p to get the number of pence
                coin_value_pence = int(coin.replace("p", ""))
                amount_paid_pence += coin_value_pence
                coins_paid += 1
            else:
                print(f"Sorry, does not accept {coin} coins.\n")

        # Give change if the user overpaid
        if amount_paid_pence > price_pence:
            change_pence = amount_paid_pence - price_pence
            # Nested condition: this if is inside another if
            if change_pence >= 100:
                print(f"Thank you for your purchase. Here is your £{change_pence/100:.2f} change.\n")
            else:
                print(f"Thank you for your purchase. Here is your {change_pence}p change.\n")
        else:
            print("Thank you for your purchase.\n")

        # Reduce stock after successful payment
        inventory[item]["stock"] -= 1
        # -= is a decrement: subtract 1 from the current value
```

```python
# Simple restock rule: if stock < 2, add 5 more
def stock_update() -> None:
    # for item in inventory iterates over the dict keys
    for item in inventory:
        if inventory[item]["stock"] < 2:
            inventory[item]["stock"] += 5
            print(f"{item} has been restocked. There are now {inventory[item]['stock']} available.")
        else:
            print(f"{item} stock is ok. There are still {inventory[item]['stock']} available.")
    print("")
```

```python
# Main program loop: pick item, pay, then update stock
while True:
    # while True creates a loop that runs until we break out
    item, is_in_stock = select_item()
    if item in exit_codes:
        print("Goodbye, Have a lovely day")
        break
    accept_payment(item, is_in_stock)
    stock_update()
```
