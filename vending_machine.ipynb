# The "Smart Vending Machine" Challenge

**GOAL:** Create a program that tracks inventory, processes payments, and dispenses
items.

**TEAM:** Pablo

```python
pound_coins = ["£2", "£1"]
penny_coins = ["50p", "20p", "10p", "5p", "2p", "1p"]
exit_codes = ["quit", "exit", "q"]
```

```python
try:
    inventory
except NameError:
    inventory = {
        "soda": {
            "stock": 10,
            "price": 1.49,
        },
        "chips": {
            "stock": 6,
            "price": 1.99,
        },
        "candy": {
            "stock": 0,
            "price": 1.19,
        }
    }
```

```python
def check_stock(item: str) -> bool:
    return inventory[item]["stock"] > 0
```

```python
def select_item() -> tuple[str, bool]:
    item = input("Select an item? ").lower().strip()
    is_in_stock = False
    if item == "soda":
        is_in_stock = check_stock(item)
    elif item == "chips":
        is_in_stock = check_stock(item)
    elif item == "candy":
        is_in_stock = check_stock(item)
    elif item in exit_codes:
        return "quit", False
    else:
        print(f"Sorry, we don't keep {item} in this machine.\n")

    if item in inventory:
            if is_in_stock:
                print(f"Certainly, we have {inventory[item]['stock']} of them.\n")
            else:
                print(f"Sorry, we are out of {item} today.\n")
    return item, is_in_stock
```

```python
def accept_payment(item: str, is_in_stock: bool) -> None:
    if is_in_stock:
        coins_paid: int = 0
        amount_paid_pence: int = 0
        price_pence = int(round(inventory[item]["price"] * 100))

        while amount_paid_pence < price_pence:
            remaining_pence = price_pence - amount_paid_pence
            remaining_display = f"£{remaining_pence/100:.2f}"
            coin = input(
                f"Please pay {remaining_display}.\n"
                f"Please enter coin {coins_paid + 1} amount (£2, £1, 50p, 20p, 10p, 5p, 2p or 1p): "
            ).strip().lower()

            if coin in exit_codes:
                print("Payment cancelled.\n")
                return

            if coin in pound_coins:
                coin_value_pence = int(coin.replace("£", "")) * 100
                amount_paid_pence += coin_value_pence
                coins_paid += 1
            elif coin in penny_coins:
                coin_value_pence = int(coin.replace("p", ""))
                amount_paid_pence += coin_value_pence
                coins_paid += 1
            else:
                print(f"Sorry, does not accept {coin} coins.\n")

        if amount_paid_pence > price_pence:
            change_pence = amount_paid_pence - price_pence
            if change_pence >= 100:
                print(f"Thank you for your purchase. Here is your £{change_pence/100:.2f} change.\n")
            else:
                print(f"Thank you for your purchase. Here is your {change_pence}p change.\n")
        else:
            print("Thank you for your purchase.\n")

        inventory[item]["stock"] -= 1
```

```python
def stock_update() -> None:
    for item in inventory:
        if inventory[item]["stock"] < 2:
            inventory[item]["stock"] += 5
            print(f"{item} has been restocked. There are now {inventory[item]['stock']} available.")
        else:
            print(f"{item} stock is ok. There are still {inventory[item]['stock']} available.")
    print("")
```

```python
while True:
    item, is_in_stock = select_item()
    if item in exit_codes:
        print("Goodbye, Have a lovely day")
        break
    accept_payment(item, is_in_stock)
    stock_update()
```
