# Unit Testing in Python
Unit Testing is the smallest level of software testing - where each unit of the system is tested individually.

The purpose is to isolate each individual part of the system to verify that they are working as specified. Typically this means testing each individual function of the code.

The use of this type of test throughout the implementation is possible to reduce the amount of bugs in the application. It works by comparing the output of a function to be tested with expected values.

## Hands On

Below we have a basic calculator class we will be using in this tutorial. We haven't officially covered a **CLASS** yet (we will do shortly in the module), but you can think of our Calculator class as a container for our functions.


In this case, our basic calculator has only four basic functions: additon, subtraction, multiplication and division.

````python
class Calculator:
    def __init__(self):
        pass

    def add(self, a, b):
        return a + b

    def sub(self, a, b):
        return a - b

    def mul(self, a, b):
        return a * b

    def div(self, a, b):
        if b != 0:
            return a / b
        else:
            return None
````

The Python's Unit Testing Framework we will be using is called `unittest`, one of the most known frameworks for testing Python code.

````python
import unittest # import the unittest library so we have access to all the unittest framework of code
````

The objective is to ensure each of our four arithmetical functions is working properly and giving the right output when receiving any two inputs.

To create a set of unit tests, we must put them all inside a **CLASS**, inheriting from `unittest.TestCase` (below).
Again, think of the class as a container - in this case holding all of our unit test functions.  And by adding `unittest.TestCase`, this tells Python that all the functions inside our class are individual unit tests...

````python
class TestCalculator(unittest.TestCase):

    def test_add(self):
        '''Test case function for addition'''
        self.calc = Calculator()
        result = self.calc.add(4, 7)
        expected = 11
        self.assertEqual(result, expected)

    def test_sub(self):
        '''Test case function for subtraction'''
        self.calc = Calculator()
        result = self.calc.sub(10, 5)
        expected = 5
        self.assertEqual(result, expected)

    @unittest.skip('Some reason')
    def test_mul(self):
        '''Test case function for multiplication'''
        self.calc = Calculator()
        result = self.calc.mul(3, 7)
        expected = 21
        self.assertEqual(result, expected)

    def test_div(self):
        '''Test case function for division'''
        self.calc = Calculator()
        result = self.calc.div(10, 2)
        expected = 4
        self.assertEqual(result, expected)
````

````python
class TestCalculator2(unittest.TestCase): # Class TestCalculator inherits the base class, TestCase, from unittest

    def test_add(self): # Any method starting with 'test' is run by unittest - self is an instance of TestCalculator
        '''Test case function for addition'''
        # ARRANGE
        self.calc = Calculator() # self (an instance of TestCalculator) now has an attribute, calc, which is an instance of the Calculator class
        num1 = 4
        num2 = 7
        expected = 11
        # ACT
        actual = self.calc.add(num1, num2) # self.calc (an instance of Calculator) calls the add method with num1 & num2 and stores the return in the variable, actual
        # ASSERT
        self.assertEqual(actual, expected, f"Expected: {expected} but got: {actual} when adding {num1} and {num2}")

    def test_sub(self):
        '''Test case function for subtraction'''
        # ARRANGE
        self.calc = Calculator()
        num1 = 10
        num2 = 5
        expected = 5
        # ACT
        actual = self.calc.sub(num1, num2)
        # ASSERT
        self.assertEqual(actual, expected, f"Expected: {expected} but got: {actual} when subtracting {num2} from {num1}")

    # @unittest.skip('Some reason')
    def test_mul(self):
        '''Test case function for multiplication'''
        # ARRANGE
        self.calc = Calculator()
        num1 = 3
        num2 = 7
        expected = 21
        # ACT
        actual = self.calc.mul(num1, num2)
        # ASSERT
        self.assertEqual(actual, expected, f"Expected: {expected} but got: {actual} when multiplying {num1} and {num2}")

    def test_div(self):
        '''Test case function for division'''
        # ARRANGE
        self.calc = Calculator()
        num1 = 10
        num2 = 2
        expected = 5.0
        # ACT
        actual = self.calc.div(num1, num2)
        # ASSERT
        self.assertEqual(actual, expected, f"Expected: {expected} but got: {actual} when dividing {num1} by {num2}")
````

We created 4 unit tests, each checking a method of the calculator class. These checks are being done through calls to `Assertions`, in this case the `assertEqual` function that verifies if the result that we expected matches the actual result.
Note that, flagging the method `test_mul` with `@unittest.skip('your_reason')` will skip the test for that method.

To simply run the tests, we use:

````python
unittest.main(argv=[''], verbosity=2, exit=False)
# .main() runs all discovered tests
# argv=[''] - argv means argument vector, a list of command line arguments - Jupyter Notebook and some IDEs automatically pass in arguments and this can cause issues with unittest so argv[''] overrides any such args with this safe empty list
# verbosity 0 is minimal, 1 is the default and shows dots (...), while 2 is detailed and shows each test name
# exit=False - Normally, when tests are run, unittest would call sys.exit() and in Jupyter Notebook, this would stop the kernal. Setting exit to False stops this behaviour.
````

````python
unittest.main(argv=[''], defaultTest='TestCalculator2', verbosity=2, exit=False)
# defaultTest specifies the class to test
````

After running you will see something like:

```
test_add (__main__.TestCalculator)
Test case function for addition ... ok
test_div (__main__.TestCalculator)
Test case function for division ... FAIL
test_mul (__main__.TestCalculator)
Test case function for multiplication ... skipped 'Some reason'
test_sub (__main__.TestCalculator)
Test case function for subtraction ... ok
```

Where addition and subtraction passed, multiplication was intentionally skipped, and division failed.

Unittest has several functions, known as `Assertions`, useful for checking the results of unit tests. Some are: `assertNotEqual(a, b)`, `assertTrue(x)`, `assertFalse(x)`, `assertIs(a, b)`, `assertIsNot(a, b)`, `assertIsNone(x)`, and much more.

# Taking it further: Organization of Test Cases

The organization of test cases (i.e. individual unit tests) is essential to make them better, more independent and help with their maintenance. For that, there are several standards related to automated tests that can be used in Unit Testing. Some of them are:

### Organizing by Instances

As seen previously, unit test cases are represented by `unittest.TestCase` instances. In other words, we created only one test class holding all the tests of its methods. In reality you might have LOTS of unit tests, and so putting all of them inside one single class might get messy and awkward.  It is possible to create different classes meaning each one can hold a bunch of specific tests for a single method.

e.g. Let's create another test case for division only.

````python
class TestCalcDiv(unittest.TestCase):

    def test_div(self):
        '''Test case function for division'''
        self.calc = Calculator()
        # tests multiple cases together - actual is passed in as the function call that gets a live return - again, the example does not show any explicit error message
        self.assertEqual(self.calc.div(10, 5), 2) # As before .assertEqual(actual, expected)
        self.assertEqual(self.calc.div(12, 2), 6)
        self.assertEqual(self.calc.div(10, 0), None)

    def test_div_error(self):
        '''Make sure ZeroDivisionError is raised when necessary'''
        self.calc = Calculator()
        self.assertRaises(ZeroDivisionError, self.calc.div, 10, 0)
        # assertRaises tests that a process raises an expected error
        # syntax: assertRaises(expected_exception, function, *args)
        # NOTE: thia test should fail since the function accounts for zero division

unittest.main(argv=[''], defaultTest='TestCalcDiv', verbosity=2, exit=False) # defaultTest specifies the test class to be run
````

The test case above will pass.

### SetUp() and TearDown()

The setUp and tearDown methods are also useful in organizing test cases. When a `setUp` function is defined, the test runner will run that method **prior** to ALL tests. Likewise, if a `tearDown` function is defined the test runner will invoke that method **after** each test.

This helps us to reduce repetitive code.  As you might notice we are creating an instance of Calculator `self.calc = Calculator()` in every unit test function. To avoid repeating code we can simply use the `setUp` function available and put this code once!

See how our division test case will look now:

````python
class TestCalcDiv(unittest.TestCase):

    def setUp(self):
        '''Set up an instance of Calculator pior to every test execution'''
        self.calc = Calculator()

    def test_div(self):
        '''Test case function for division'''
        self.assertEqual(self.calc.div(10, 5), 2)
        self.assertEqual(self.calc.div(12, 2), 6)

    def test_div_error(self):
        '''Make sure ZeroDivisionError is raised when necessary'''
        self.assertRaises(ZeroDivisionError, self.calc.div, 10, 0)

unittest.main(argv=[''], defaultTest='TestCalcDiv', verbosity=2, exit=False)
````
