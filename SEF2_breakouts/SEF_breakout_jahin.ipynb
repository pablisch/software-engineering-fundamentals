```python
class ItemsStock:
    """Class used to represent items, their stock and their price"""

    def __init__(self):

        # "item_name":
        # {
        #     "stock": stock,
        #     "price": price
        # }
        self._items_dict = {}

    def add_item(self, item_name, stock, price) -> None:
        assert isinstance(stock, int) and stock >= 0, "Stock must be a non-negative integer"
        assert isinstance(price, (int, float)) and price >= 0, "Price must be a non-negative number"
        assert item_name == item_name.capitalize(), "Item name must be capitalized"
        assert item_name not in self._items_dict, "Item already exists in the inventory"
        
        if isinstance(price, int):
            price = float(price)

        self._items_dict[item_name] = {'stock': stock, 'price': price}
        
    def is_item(self, item_name) -> bool:
        return item_name in self._items_dict
    
    def get_items(self) -> list[str]:
        return list(self._items_dict.keys())

    def is_item_available(self, item_name) -> bool:
        assert (item_name in self._items_dict), "Item not found in the inventory"
        return self._items_dict[item_name]['stock'] > 0
    
    def get_available_items(self) -> list[str]:
        return [item for item, details in self._items_dict.items() if details['stock'] > 0]
        
    def get_item_price(self, item_name) -> float:
        assert (item_name in self._items_dict), f"Item {item_name} not found in the inventory"
        return self._items_dict[item_name]['price']
    
    def reduce_stock_by_one(self, item_name) -> None:
        assert (item_name in self._items_dict), "Item not found in the inventory"
        assert self._items_dict[item_name]['stock'] > 0, "Item is out of stock"
        self._items_dict[item_name]['stock'] -= 1

    def set_machine_slots(self, slots: list)-> None:
        """
        Slots are Soda, Chips, Candy
        """
        self.slots = slots
        
    def scan_restock(self, threshold: int = 2) -> None:
        """
        For Loop:
        Iterate through slots and give stock alerts 
        """

        for item_name, item in self._items_dict.items():
            qty = item["stock"]

            if qty < threshold:
                print(f"Restock Needed- {item_name} | Stock: {qty}") 
            else:
                print(f"Slot {item_name} OK | Stock: {qty}")


def get_item_input(item_stock: 'ItemsStock') -> str:

    item = input("Enter Item name: ").capitalize()

    item_names = item_stock.get_items()
    items_in_stock = item_stock.get_available_items()

    while item not in items_in_stock:

        if item in item_names: # Valid item but out of stock
            print(f"{item} is out of stock! Enter 'Help' for a list of available items.")
        elif item == "Help":
            print(f"Available items: {', '.join(items_in_stock)}")
        else:
            print("No item of that name! Enter 'Help' for a list of available items.")
        
        item = input("Enter Item name: ").capitalize()
    
    return item
```

```python
item_stock = ItemsStock()
item_stock.add_item("Candy", 10, 2.0)
item_stock.add_item("Chips", 5, 0.79)
item_stock.add_item("Soda", 8, 2.0)
```

```python
item_input = get_item_input(item_stock)
```

```python
print(f"Item Price : {item_stock.get_item_price(item_input)} (no change given!)")

total = 0.0
item_price = item_stock.get_item_price(item_input)

while (total < item_price):
    total += float(input("Input coin: "))
    
item_stock.reduce_stock_by_one(item_input)
print(f"Total inserted: {total:.2f}\nThank you for your purchase of {item_input}!")
```
