```python
stock = {"soda": 1, "chips": 5, "candy": 0}

user_request = input("What would you like to buy")

if user_request == "soda":
    if stock["soda"] > 0:
        print("soda in stock")
    
    else:
        print("soda not in stock")

elif user_request == "chips":
    if stock["chips"] > 0:
        print("chips in stock")
    
    else:
        print("chips not in stock")

elif user_request == "candy":
    if stock["candy"] > 0:
        print("candy in stock")
    
    else:
        print("candy not in stock")

else:
    print(f"unknown input {user_request}, please enter soda chips or candy")
```

```python
for item, qty in stock.items():
    if qty < 2:
        print(f"{item} needs restocking!")
```

```python
prices = {"soda": 1.00, "chips": 1.50, "candy": 0.50}
coins = 0
price = prices[user_request]

while (coins <= price):
    amount = float(input("Enter coins input:"))
    coins += amount

print(f"paid {amount} your change is {price-amount}")
print(f"enjoy your {user_request}")
```
