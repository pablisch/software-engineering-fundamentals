```python
# Team members: Joshua, Elliott, Hamish

# List of machine slots (item, quantity)
machine_slots = [
    ("Soda", 10),
    ("Chips", 10),
    ("Candy", 10)
]

# Inventory check
item = input()
if item in machine_slots:
  # Find tuple related to item
  for slot in machine_slots:
    if slot[0] == item:
      print("Item is in stock!" if slot[1] > 0 else "Item is out of stock :(")
else:
  print("Please enter a valid item  ")
```

```python
# Payment loop
amount = 0
price = 17

print(f"Item price: {price}")

while amount < price:
  accepted_coins = [1, 2, 5]
  print(f"\nAmount entered: {amount}")
  coin = int(input("Please enter some coins. (1, 2, 5)\n"))
  if coin in accepted_coins:
    amount += coin
  else:
    print("\nUnknown coin entered. Please try again...")

print("\nItem purchased! Enjoy :)")
```

```python
# Update stock

for item, quantity in machine_slots:
    if quantity < 2:
        print(f"Restock needed: {item} (only {quantity} left)")
```
