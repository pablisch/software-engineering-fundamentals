```python
# --------- Part 1 ----------

stock = {
    "Soda" : 3,
    "Chips" : 10,
    "Candy" : 10
}

def check_stock(item: str) -> bool:
  if stock[item] <= 0:
    return False
  return True
```

```python
# --------- Part 2 ----------
prices = {
    "Soda" : 1.25,
    "Chips" : 1.50,
    "Candy" : 2.00
}

def get_money(item: str) -> int:
  total = 0
  while prices[item] > total:
    total += float(input("Enter Money: "))
  stock[item] -= 1
  return total - prices[item]
```

```python
# --------- Part 3 ----------

def update_stock():
  for item in stock.keys():
    if stock[item] < 2:
      print(f"{item} needs to be restocked")
```

```python
# --------- Final ----------
def loop():
  product = input("What you want? ")
  if product not in stock.keys():
    print("Invalid item")
    return
  if not check_stock(product):
    print("Out of stock")
    return
  print(f"£{get_money(product)} change")
  update_stock()

while True:
  loop()
```
