Inventory Check (if/elif/else)

```python
items = {
    "Coke":3,
    "Redbull":2,
    "Fanta":5,
    "Pepsi":6,
    "Sprite":8,
}

choice = input("Select a drink (Coke, Redbull, Fanta, Pepsi, Sprite): ")

if choice == "Coke":
    index = 0
    price = 1.90
elif choice == "Redbull":
    index = 1
    price = 1.65
elif choice == "Fanta":
    index = 2
    price = 1.90
elif choice == "Pepsi":
    index = 3
    price = 1.90
elif choice == "Sprite":
    index = 4
    price = 1.90
else:
    print("Drink not found")
    index = -1

if index != -1:
    name, num = list(items.items())[index]

    if num > 0:
        print(name, "is in stock")
    else:
        print(name, "is out of stock")
```

Extension (The "Secret Discount")

```python
coupon = input("Enter coupon code (or press enter to skip): ")

if coupon == "DISCOUNT20":
  price = price * 0.8
  print("Discount applied. New price is £", price)

else:
  print("No discount applied")
```

Stock Update (for loop)

```python
def stock_update(choice):
    for name, qty in items.items():
        if name == choice:
            items[name] -= 1
            qty = items[name]

            if qty < 2:
                items[name] += 5
                print(f"{name} had {qty} remaining, restocked it to {items[name]}.")
```

Payment Loop (while loop)

```python
# Payment Loop
total_paid = 0

while total_paid < price:
  coin = float(input("Insert coin (£): "))
  total_paid = total_paid + coin
  print("Total paid so far: £", total_paid)

change = total_paid - price
print("Payment complete")
stock_update(choice)

if change >0:

  print("Your change is £", round(change, 2))
```
