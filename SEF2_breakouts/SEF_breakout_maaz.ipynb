```python
 #inventory
soda_quantity = 5
chips_quantity = 10
candy_quantity = 15

item = input("What would you like? Soda, Chips or Candy").lower()

selected_item_name = None

if item == "soda":
  if soda_quantity > 0:
    price = 1.50
    print(f"Soda is in stock. Price: £{price}")
    selected_item_name = "soda"
  else:
    price = None
    print("Soda is out of stock")

elif item == "chips":
  if chips_quantity > 0:
    price = 1.00
    print(f"Chips is in stock. Price: £{price} ")
    selected_item_name = "chips"
  else:
    price = None
    print("Chips is out of stock")

elif item == "candy":
  if candy_quantity > 0:
    price = 2.50
    print(f"Candy is in stock. Price: £{price} ")
    selected_item_name = "candy"
  else:
    price = None
    print("Candy is out of stock")

else:
  price = None
  print("Invalid item selected. ")

# Payment Loop

if price is not None:
    total_paid = 0.00

    while total_paid < price :
      coin = float(input(" INSERT COINS: "))
      total_paid += coin
      print(f" Total paid so far: ${total_paid:.2f} ")

    print(f" Your payment is done. The total amount paid is ${total_paid:.2f} ")

    if total_paid > price:
      change = total_paid - price
      print(f" Your change: ${change:.2f} ")

    if selected_item_name == "soda":
        soda_quantity -= 1
    elif selected_item_name == "chips":
        chips_quantity -= 1
    elif selected_item_name == "candy":
        candy_quantity -= 1

    print("Item dispensed. Thank you! ")

else:
  print("Item unavailable. ")

# stock checks

if soda_quantity <=2:
  print("Soda is running low needs to be restocked")

if chips_quantity <=5:
  print("Chips is running low needs to be restocked")

if candy_quantity <=6:
  print("candy is running low needs to be restocked")

```
