# Simple testing for PART 1

Here are some simple techniques that you can chose to use to test your Python code for PART 1 of the assessment:

*   print()
*   type()
*   len()
*   isinstance()
*   assert

```python
# 1. Define a piece of data
fruits = ["apple", "banana", "cherry"]

# 2. print() - View the data in the console
print(f"Current list: {fruits}")

# 3. type() - Check the exact class of the object
print(f"The data type is: {type(fruits)}")

# 4. len() - Check the size of the object
list_length = len(fruits)
print(f"Number of items: {list_length}")

# 5. isinstance() - A flexible way to check if data matches a type (returns True/False)
is_it_a_list = isinstance(fruits, list)
print(f"Is it a list? {is_it_a_list}")

# 6. assert() - Test a condition; if it's False, the program will crash with an error
# This is used to "guard" your code against bad data.
assert len(fruits) == 3, "Error: The list should have exactly 3 items!"
assert isinstance(fruits, list), "Error: Data must be a list!"

print("All tests passed successfully!")
```

# Key Differences at a Glance

Method 	Purpose	Best Used For...

*   **print()**	Display	Seeing what a variable holds during execution.
*   **type()**	Identity	Seeing exactly what an object is (e.g., <class 'list'>).
*   **len()**	Measurement	Counting items in lists, strings, or dictionaries.
*   **isinstance()**	Verification	Checking if a variable is a certain type (supports inheritance).
*   **assert**	Validation	Hard-stopping the code if a condition isn't met (debugging).

# Important Note on assert

Unlike the other four, *assert* is a statement, not a *function*.

You usually don't see it used with parentheses like assert(x == y), though Python allows it.

Its primary job is to act as a "sanity check" while you are developing your code.

#Example

In this example, we’ll create a function that processes a username.

This function will use all five methods to "guard" the code, ensuring the input is a string and meet certain length requirements before proceeding.

```python
def process_username(user_input):
    # 1. print() - Log the received input for debugging
    print(f"\n--- Processing input: '{user_input}' ---")

    # 2. type() - Check the exact class (useful for logging)
    input_type = type(user_input)
    print(f"DEBUG: Input class is {input_type}")

    # 3. isinstance() - The "gold standard" for type-checking in logic
    if not isinstance(user_input, str):
        print("ERROR: Input is not a string! Rejecting.")
        return  # Exit early if type is wrong

    # 4. len() - Check if the string meets a minimum length requirement
    if len(user_input) < 3:
        print(f"ERROR: '{user_input}' is too short ({len(user_input)} chars).")
        return

    # 5. assert() - Final "sanity check" before critical logic
    # If this fails, it means there's a bug in our code logic above.
    assert isinstance(user_input, str) and len(user_input) >= 3, "Validation logic failed!"

    print(f"SUCCESS: Username '{user_input}' is valid and ready for the database.")

# --- Testing the function with different data ---
process_username("JD")          # Fails len() check
process_username(12345)        # Fails isinstance() check
process_username("Skywalker")  # Passes all checks
```

# **When to use which?**

*   **isinstance()** is preferred over type() for logic because it also handles subclasses (e.g., checking if something is a "number" by checking against int or float).
*   **assert** is a debugging tool for developers to catch "impossible" states. It is not recommended for production-level user validation because it can be globally disabled in Python for performance (using the -O flag), which would skip your checks entirely.
*   **len()** is the standard way to prevent "empty" or "overly large" data from crashing your system.
*   **print()** is your primary tool for "trace" debugging—seeing exactly what value and type passed through a function at a specific moment.


If you're building a real app, you can use try...except blocks to handle these errors more gracefully so the program doesn't crash

# **Key Components of this Workflow**

* try: Wraps the "risky" code where you expect potential errors.
* raise: Manually triggers an exception. This is better than assert for user input because it won't be disabled in production.
* except: The safety net that executes only if an error occurs. Specifying the error type (like ValueError) is a best practice to avoid hiding unexpected bugs.
* finally: Used for "cleanup" actions, like closing a database connection or a file, ensuring they happen regardless of errors.

# **Pro-Tip: Assertions vs. Exceptions**

Use assert for internal logic (e.g., "This list I created myself should never be empty"). It is for catching your own bugs during development.


Use try...except for external factors (e.g., user input, network issues, or files). These handle anticipated failures in a running program

```python
def safe_process_username(user_input):
    try:
        # 1. Type validation
        if not isinstance(user_input, str):
            # We "force" a TypeError if it's not a string
            raise TypeError(f"Expected a string, but got {type(user_input).__name__}")

        # 2. Length validation
        if len(user_input) < 3:
            # We "force" a ValueError if the content is invalid
            raise ValueError(f"Username '{user_input}' is too short (min 3 chars)")

        # 3. Success logic (The "Happy Path")
        print(f"SUCCESS: {user_input} has been registered!")

    except (TypeError, ValueError) as error:
        # 4. Handle the errors gracefully
        print(f"LOGGED ERROR: {error}")
        print("Please try again with a valid string of at least 3 characters.")

    finally:
        # 5. Cleanup (Always runs, success or failure)
        print("Verification process complete.")

# --- Testing the 'Safe' function ---
safe_process_username(12345)        # Catches TypeError
safe_process_username("JD")         # Catches ValueError
safe_process_username("Skywalker")  # Runs Success logic
```

# Advanced using a Loop
To keep asking the user for input until they provide a valid response, you wrap your validation logic in a while **True loop**.


This creates an infinite loop that only stops when it hits a break statement or a return

```python
def get_valid_age():
    while True:  # 1. Start an infinite loop
        user_input = input("Please enter your age: ")

        try:
            # 2. Try to convert and validate the input
            age = int(user_input)

            if age < 0 or age > 120:
                # Manually trigger an error if the range is wrong
                raise ValueError("Age must be between 0 and 120.")

            # 3. If we get here, the input is valid!
            # break (or return) exits the loop
            return age

        except ValueError as e:
            # 4. If an error happened, print it and the loop restarts
            print(f"Invalid input: {e}. Please try again.")

# Call the function
final_age = get_valid_age()
print(f"Verified age: {final_age}")
```

# **Why this works:**

**The Loop:** The while True statement ensures the code block repeats indefinitely.

**The Trap:** The try...except block "traps" any errors (like someone typing "twenty" instead of "20").

**The Restart:** When an error is caught in the except block, the code below it in the try block is skipped, and the loop naturally starts over from the top.

**The Exit:** The break or return statement is placed at the very end of the try block. It only runs if no errors occurred in the lines above it.


# **Best Practice: Specific Exceptions**

Always catch specific errors like ValueError rather than using a "bare" except:. This prevents you from accidentally hiding other bugs, like a typo in your variable names.
