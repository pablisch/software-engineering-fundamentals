# Typing in python

Strong Typing:
Python is considered a strongly typed language because it enforces strict type checking during operations. Variables in Python have a specific data type, and the interpreter checks for compatibility when performing operations. If there is an attempt to perform an operation with incompatible data types, Python will raise a TypeError.

Example of strong typing in Python:

```python
x = 10
y = "20"

# This line would raise a TypeError since Python is strongly typed.
result = x + y
```

Dynamic Typing:
Python

Example of dynamic typing in Python:

```python
x = 10
print(type(x))  # Output: <class 'int'>

x = "Hello"
print(type(x))  # Output: <class 'str'>
```

In the example above, the variable x changes its data type from an integer to a string without any explicit type declaration.

The combination of strong typing and dynamic typing in Python provides a balance of flexibility and safety. It allows for a more natural coding experience and faster development, as you don't need to worry about type declarations, while still catching potential type-related errors at runtime. However, this also means that developers need to be cautious when performing operations on variables with different data types to avoid unexpected behavior or errors.

Boolean Operators -
**Comparison Operators:**
Comparison operators compare two values and return a boolean value based on the comparison.

```python
# Comparison operators
u = 5
v = "5"
x = 5
y = 10
a = 10
b = 10
c = 1000
d = 1000

print(u == v)  # Output: False (u is not equal to v because they are not the same type)
print(x == y)  # Output: False (x is not equal to y)
print(a == b)  # Output: True (a is equal to b)
print(a is b)  # Output: True (a is the same object as b - works for small ints from -5 to 256 as they are pre-created)
print(c is d)  # Output: False (c is not the same object as d)
print(c is c)  # Output: False (c is the same object as c)
print(x != y)  # Output: True (x is not equal to y)
print(x > y)   # Output: True (x is greater than y)
print(x < y)   # Output: False (x is less than y)
print(x >= y)  # Output: True (x is greater than or equal to y)
print(x <= y)  # Output: False (x is less than or equal to y)
```

**Logical Operators**
Logical operators perform logical operations on boolean values and return a boolean result.

```python
# Logical operators
a = True
b = False

print(a and b)  # Output: False (both a and b are not True)
print(a or b)   # Output: True (at least one of a or b is True)
print(not a)    # Output: False (not True is False)
```

# **Logic (Conditional Statements)**

In Python, if and else are conditional statements that allow you to control the flow of your program based on certain conditions. They are essential for decision-making and executing specific blocks of code depending on whether a condition is true or false. Let's dive into the tutorial on using if and else statements in Python:

**if Statement:**
The if statement is used to execute a block of code only if a specified condition is true. The general syntax is as follows:

```python
if condition:
    # Code block to be executed if the condition is true
```

The if statement is indented and will only execute if the condition evaluates to True. If the condition is False, the block is skipped, and the program moves on to the next line of code.

Example:

```python
# If statement example
num = 10

if num > 0:
    print("The number is positive.")
```

**else Statement:**
Theelsestatementif condition is false. The general syntax is as follows:

```python
if condition:
    # Code block to be executed if the condition is true
else:
    # Code block to be executed if the condition is false
```

 If the if condition is true, the else block is skipped. However, if the if condition is false, the else block is executed.

Example:

```python
# If-else statement example
num = -5

if num > 0:
    print("The number is positive.")
else:
    print("The number is not positive.")
```

**elif Statement:**
The elif (short for "else if") statement is used to check additional conditions after the initial ifstatement.

```python
if condition1:
    # Code block to be executed if condition1 is true
elif condition2:
    # Code block to be executed if condition2 is true
else:
    # Code block to be executed if all conditions are false
```

Use elif blocks to handle different cases. Once one of the conditions evaluates to True, the corresponding block of code is executed, and the rest of the if-elif-else statement is skipped.

```python
# If-elif-else statement example
num = 0

if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

Conditional statements like if, else, and elif are fundamental to making decisions in Python programs. They allow you to control the flow of your code based on specific conditions, making your programs more dynamic and responsive. By mastering these logical constructs, you can create powerful and versatile Python applications

# **Loops**

In Python, loops are used to execute a block of code repeatedly based on a specified condition. There are two main types of loops in Python: for loops and while loops. Let's explore how to use both types of loops in this Python tutorial on loops:

**1. for Loop:**
The for loop is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block of code for each element in the sequence. The general syntax of a for loop is as follows:

```python
for variable in sequence:
    # Code block to be executed for each element in the sequence
```

The loop variable takes the value of each element in the sequence during each iteration.

Example:

```python
# for loop example
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

**2. while Loop:**
The while loop is used to repeatedly execute a block of code as long as a specified condition is true. The general syntax of a while loop is as follows:

```python
while condition:
    # Code block to be executed as long as the condition is true
```

The loop will continue to execute the code block until the condition evaluates to False. Be cautious not to create infinite loops, where the condition is always true, leading to an endless loop.

Example:

```python
# while loop example
count = 0

while count < 5:
    print("Current count:", count)
    count += 1
```

**break and continue Statements:**
The break statement is used to exit a loop prematurely when a certain condition is met. It immediately terminates the loop, and the program continues to execute the next line of code after the loop.
The continue statement is used to skip the rest of the current iteration and move on to the next iteration of the loop. It does not exit the loop but skips the remaining code for the current iteration.
Example:

```python
# break and continue statements example
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    if num == 3:
        continue  # Skip printing 3 and continue to the next iteration
    elif num == 5:
        break     # Exit the loop when 5 is encountered
    print(num)
```

Loops are powerful constructs in Python that allow you to repeat tasks efficiently and handle sequences of data. Whether you're iterating over a collection with for loops or executing code based on a condition with while loops, understanding how to use loops is essential for solving various programming problems in Python. Happy coding with loops!

```python
# Using the range() function with a for loop
for num in range(1, 6):
    print(num)

```

```python
# Using a for loop to find the sum of numbers in a list
numbers = [10, 20, 30, 40, 50]
sum_of_numbers = 0

for num in numbers:
    sum_of_numbers += num

print("Sum of numbers:", sum_of_numbers)
```

```python
# Using a while loop to reverse a string
word = "Python"
reversed_word = ""
index = len(word) - 1

while index >= 0:
    reversed_word += word[index]
    index -= 1

print("Reversed word:", reversed_word)
```

```python
# Using the range() function with a step in a for loop
for num in range(1, 11, 2):
    print(num)
```

```python
# Using continue to skip even numbers in a loop
for num in range(1, 11):
    if num % 2 == 0:
        continue
    print(num)
```

```python
# Using nested loops to multiply elements of two matrices
matrix1 = [[1, 2], [3, 4]]
matrix2 = [[5, 6], [7, 8]]
result = [[0, 0], [0, 0]]

for i in range(len(matrix1)):
    for j in range(len(matrix2[0])):
        for k in range(len(matrix2)):
            result[i][j] += matrix1[i][k] * matrix2[k][j]

print("Matrix multiplication result:")
for row in result:
    print(row)
```
